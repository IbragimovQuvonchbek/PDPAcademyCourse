```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
total_passengers = len(df)
male_passengers = len(df[df['Sex'] == 'male'])
female_passengers = len(df[df['Sex'] == 'female'])
print(f"Barcha yo'lovchilar soni: {total_passengers}")
print(f"Erkaklar soni: {male_passengers}")
print(f"Ayollar soni: {female_passengers}")
```

```python
labels = ['Erkaklar', 'Ayollar']
sizes = [male_passengers, female_passengers]
colors = ['blue', 'pink']
explode = (0.1, 0)
```

```python
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.axis('equal')  # Pie chartni aylana shaklida qilish uchun
plt.title('Erkaklar va Ayollar soniga mos pie chart')
plt.show()
```

```python
total_fare = df['Fare'].sum()
average_fare = df['Fare'].mean()

print(f"Biletlari narxining umumiy summasi: {total_fare}")
print(f"Bitta bilet narxining o'rtacha qiymati: {average_fare}")
```

```python
age_15_40 = len(df[(df['Age'] >= 15) & (df['Age'] <= 40)])
percentage_15_40 = (age_15_40 / total_passengers) * 100

print(f"15-40 yosh oralig'idagi yo'lovchilar foizi: {percentage_15_40:.2f}%")
```

```python
total_children = len(df[df['Age'] < 16])
survived_children = len(df[(df['Age'] < 16) & (df['Survived'] == 1)])
percentage_survived_children = (survived_children / total_children) * 100 if total_children > 0 else 0

print(f"Tirik qolgan bolalar foizi: {percentage_survived_children:.2f}%")
```
